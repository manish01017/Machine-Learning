## Named Entity Recognition

Named Entity Recognition is an NLP technique used to identify and classify key information (entities) in text, such as names of people, places, organizations, dates, and more.
       
NER helps extract structured data from unstructured text, making it easier to analyze and use in downstream applications like search, information retrieval, and question answering.
