# Spam Identifier model

```python
import pandas as pd

data = pd.read_csv('SMSSpamCollection', sep='\t', names=['label', 'message'])
```

```python
data
```

```python
import nltk

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
import re
```

```python
from nltk import WordNetLemmatizer

lm=WordNetLemmatizer()
```

```python
corpus=[]

for i in range(len(data)):
    msg=re.sub('[^a-zA-Z]',' ', data['message'][i])
    review=[lm.lemmatize(word.lower()) for word in msg.split() if not word in set(stopwords.words('english'))]
    rev=' '.join(review)
    corpus.append(rev)
    
    
  

```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#max_feature : features selection for bag of words
#ngrams : feature creation by using multiple columns
#binary : vectorizer produced binary vector (0 and 1)


vectorizer=CountVectorizer(max_features=100 ,ngram_range=(2,2), binary=True)
X=vectorizer.fit_transform(corpus)
X
```
